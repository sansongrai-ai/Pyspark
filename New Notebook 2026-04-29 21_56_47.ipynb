```python
%python 
data = [
    (1, "laptop", 1000),
    (2, "phone", 500),
    (3, "watch", 100),
    (4, "charger", 50),
    (5, "headphones", 100),
    (6, "mouse", 50),
    (7, "keyboard", 100)
]

columns = ["id", "product", "price"]

df = spark.createDataFrame(data, columns)
df.show()
```

```python
df.filter(df.price > 100).show()
```

```python
%python 

from pyspark.sql.functions import sum

data1 = [
    (1, "laptop", 1000),
    (2, "phone", 500),
    (3, "watch", 100),
    (4, "charger", 50),
    (5, "headphones", 100),
    (6, "mouse", 50),
    (7, "keyboard", 100)
]

columns1 = ["id", "product", "price"]

df = spark.createDataFrame(data1, columns1)

result = df.groupBy("product").agg(sum("price").alias("Total_sales"))

result.show()
```

```python
result = df.orderBy(df.price.desc())
result.show()
```

```python
from pyspark.sql.functions import col

df = spark.createDataFrame(data1, columns1)

new_result = df.withColumn("Discounted_price", col("price") - 100)

new_result.show()
```

```python
df = spark.createDataFrame(data1, columns1)

drop_df = df.dropDuplicates(["product"])

drop_df.show()
```
