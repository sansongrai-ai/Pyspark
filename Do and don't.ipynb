For DO'S
- 1. Filtertion early(reduce data as fast as it can)
df=df.filter(col("event_date")>=2023-04-31)

2. Select only required columns:

- df=df.select("user_id")

3. Use proper partition:
df.write.partitionBy("event_date").parquet("path")
read - df=spark.read.parquet("path").filter("event_date = '2012-05-06' ")

4. Use Broadcast Join: Broadcast for small tables

5. Handle bad data explicitly:
   spark.read.option("mode","permissive mode")
   - store corrupt data separately

7. Handles skew data:  


For DON'T

 - df.collect()

2. Don't use repartition:

4. Don't ignore NULL handiling because NULL can breaks aggregation silently.

P1- 10 million of rows data
P2 - 5rows

Transformation:
 1. Narrow: Each output partition depends only on one input partition.
 
   - no suffle happens.
   - df=spark.read.parquet("orders")
   - df2 =df.filter(df.amount >100)
   - df3 =df2.withColumn("amount_tax" , df.amount*1.18)
   - No data exchange between partition.
   - Example: filter, select, withColumn
   - Faster than wide 


 2.  Wide: Where data from multiple partitions must be combined.

     - It causes suffle.
     - There will be data movement across nodes.
     - df.groupBy("country")
     - Data is resuffled across your clusters.
     - Eg: groupBy, re-partition, Join, orderBy
     - Expensive and slow.
