```python
data_emp = [
    (1, "John", "Doe", 60000, 1),
    (2, "Jane", "Smith", 80000, 2),
    (3, "Sam", "Brown", 75000, 1),
    (4, "Lisa", "Ray", 90000, 2),
    (5, "Tom", "Hanks", 50000, 1)
]


data_dept = [
    (1, "marketing"),
    (2, "engineering")
]
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql import functions as F
spark = SparkSession.builder.appName("salaryDifference").getOrCreate()
```

```python
emp_columns = ["emp_id", "first_name", "last_name", "salary", "dept_id"]
dept_columns = ["dept_id", "dept_name"]

db_employee = spark.createDataFrame(data_emp, emp_columns)
db_dept = spark.createDataFrame(data_dept, dept_columns)
```

```python
df = db_employee.join(db_dept, "dept_id")
```

```python
max_salary = df.groupBy("dept_name").agg(
    F.max("salary").alias("max_salary")
)
```

```python
marketing_salary = max_salary.filter(
    F.col("dept_name") == "marketing"
).select("max_salary").collect()[0][0]

engineering_salary = max_salary.filter(
    F.col("dept_name") == "engineering"
).select("max_salary").collect()[0][0]


```

```python
display(marketing_salary-engineering_salary)
```
