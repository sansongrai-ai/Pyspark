```python
from pyspark.sql import SparkSession
from  pyspark.sql.types import *
from pyspark.sql.functions import *
```

```python
#Define schema manually
covid_schema=StructType([
    StructField("Province/State", StringType(), True),
    StructField("Country/Region", StringType(), True),
    StructField("Lat", DoubleType(), True),
    StructField("Long", DoubleType(), True),
    StructField("Date", DateType(), True),
    StructField("Confirmed", IntegerType(), True),
    StructField("Deaths", IntegerType(), True),
    StructField("Recovered", IntegerType(), True),
    StructField("Active", IntegerType(), True),
    StructField("WHO Region", StringType(), True)
])
 
```

```python
df = spark.read.format ("csv") \
    .option ("header", "true") \
    .schema(covid_schema) \
    .load("/Volumes/workspace/default/pyspark/covid_19_clean_complete.csv")



```

```python
df = df.withColumnRenamed ("WHO Region", "WHO_Region")
```

```python
df = df.dropDuplicates([
    "Province/State",
    "Country/Region",
    "Date"
    ])
```

```python
df = df.fillna ({"Province/State": "N/A"})
display(df)
```

```python
df = df.fillna ({"WHO_Region": "N/A" })
display(df)
```

```python
df = df.withColumn ("Confirmed", coalesce(col("Confirmed"), lit(0)))

df = df.withColumn ("Deaths", coalesce(col("Deaths"), lit(0)))

df = df.withColumn ("Recovered", coalesce(col("Recovered"), lit(0)))
```

```python
df = df.filter((col("lat") !=0) & (col("long") !=0))
```

```python
df = df.filter(
    (col("Confirmed") >=0) &
    (col("Deaths") >=0) &
    (col("Recovered") >=0)
)
```
